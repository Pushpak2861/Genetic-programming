```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
def scal(data):

  X = data.copy()
  # Columns for each scaling method
  standard_cols = ['BMI', 'Age', 'MentHlth', 'PhysHlth']
  minmax_cols = ['GenHlth', 'Education', 'Income']
  binary_cols = [col for col in X.columns if col not in standard_cols + minmax_cols]

  # Initialize scalers
  standard_scaler = StandardScaler()
  minmax_scaler = MinMaxScaler()
  robust_scaler = RobustScaler()

  # Apply scaling
  X_standard_scaled = standard_scaler.fit_transform(X[standard_cols])
  X_minmax_scaled = X[minmax_cols]
  X_binary = X[binary_cols]  # No scaling for binary columns

  # Combine scaled data
  import numpy as np
  X_scaled = np.hstack([X_standard_scaled, X_minmax_scaled, X_binary])

  # Convert back to DataFrame for readability
  scaled_columns = standard_cols + minmax_cols + binary_cols
  X_scaled_df = pd.DataFrame(X_scaled, columns=scaled_columns)

  print(X_scaled_df.head())

  return X_scaled_df

```

The fitness function here is the percentage of outputs wrongly predicted.

You can write your own fitness function if you prefer.

```python
from sklearn.metrics import accuracy_score, log_loss
import numpy as np

def fitness_eval(individual, points):
    high_variance_threshold = 0.1
    x = points[0]
    y = points[1]
    penalty = 0

    if individual.invalid:
        return np.NaN,

    try:
        pred = eval(individual.phenotype)
    except (FloatingPointError, ZeroDivisionError, OverflowError, MemoryError):
        return np.NaN,

    assert np.isrealobj(pred)

    try:
        Y_class = [1 if pred[i] > 0 else 0 for i in range(len(y))]
        pred = [int(p) for p in pred]
        Y_prob = [1 / (1 + np.exp(-p)) for p in pred]
    except (IndexError, TypeError):
        return np.NaN,

    accuracy = accuracy_score(y, Y_class)
    logloss_value = log_loss(y, Y_prob)

    # Penalize for high log loss
    max_logloss_threshold = 0.5
    if logloss_value > max_logloss_threshold:
        penalty += 1.0 * (logloss_value - max_logloss_threshold)

    # Variance Penalty
    y_class_variance = np.var(Y_class)
    if y_class_variance > high_variance_threshold:
        penalty += 0.5 * (y_class_variance - high_variance_threshold)

    # Combine accuracy and penalties
    fitness = accuracy - penalty / 5  # Adjusted scaling of penalty

    # Ensure fitness is not negative
    fitness = max(fitness, 0.1)  # Provide a small positive value

    return (fitness,)
```

Set the Grammatical Evolution parameters.

Make sure you set a random seed just in case we need to re-run your experiments.

```python
POPULATION_SIZE = 1000
MAX_GENERATIONS = 100
P_CROSSOVER = 0.5
P_MUTATION = 0.03
ELITE_SIZE = 1
HALL_OF_FAME_SIZE = 1

TOURNAMENT_SIZE = 6
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

CODON_CONSUMPTION = 'lazy'
GENOME_REPRESENTATION = 'list'
MAX_GENOME_LENGTH = None

MAX_INIT_TREE_DEPTH = 12
MIN_INIT_TREE_DEPTH = 3
MAX_TREE_DEPTH = 90
MAX_WRAPS = 0
CODON_SIZE = 255

REPORT_ITEMS = ['gen', 'invalid', 'avg', 'std', 'min', 'max',
                'best_ind_length', 'avg_length',
                'best_ind_nodes', 'avg_nodes',
                'best_ind_depth', 'avg_depth',
                'avg_used_codons', 'best_ind_used_codons',
                'structural_diversity', 'fitness_diversity',
                'selection_time', 'generation_time']
```

Create a toolbox.

```python
toolbox = base.Toolbox()

# define a single objective, minimising fitness strategy:
creator.create("FitnessMax", base.Fitness, weights=(1.0,))

creator.create('Individual', grape.Individual, fitness=creator.FitnessMax)


toolbox.register("populationCreator", grape.sensible_initialisation, creator.Individual)

toolbox.register("evaluate", fitness_eval)

# Tournament selection:
toolbox.register("select", tools.selTournament, tournsize=TOURNAMENT_SIZE)

# Single-point crossover:
toolbox.register("mate", grape.crossover_onepoint)

# Flip-int mutation:
toolbox.register("mutate", grape.mutation_int_flip_per_codon)
```

```python
# create initial population (generation 0):
population = toolbox.populationCreator(pop_size=POPULATION_SIZE,
                                           bnf_grammar=BNF_GRAMMAR,
                                           min_init_depth=MIN_INIT_TREE_DEPTH,
                                           max_init_depth=MAX_INIT_TREE_DEPTH,
                                           codon_size=CODON_SIZE,
                                           codon_consumption=CODON_CONSUMPTION,
                                           genome_representation=GENOME_REPRESENTATION
                                            )

# define the hall-of-fame object:
hof = tools.HallOfFame(HALL_OF_FAME_SIZE)

# prepare the statistics object:
stats = tools.Statistics(key=lambda ind: ind.fitness.values)
stats.register("avg", np.nanmean)
stats.register("std", np.nanstd)
stats.register("min", np.nanmin)
stats.register("max", np.nanmax)
```

```python
for ind in population:
  ind.fitness.values = toolbox.evaluate(ind, points=[X_train, y_train])
  print(ind.fitness.values)
```

Run Grammatical Evolution.

```python
population, logbook = algorithms.ge_eaSimpleWithElitism(population, toolbox, cxpb=P_CROSSOVER, mutpb=P_MUTATION,
                                              ngen=MAX_GENERATIONS, elite_size=ELITE_SIZE,
                                              bnf_grammar=BNF_GRAMMAR,
                                              codon_size=CODON_SIZE,
                                              max_tree_depth=MAX_TREE_DEPTH,
                                              max_genome_length=MAX_GENOME_LENGTH,
                                              points_train=[X_train, y_train],
                                              codon_consumption=CODON_CONSUMPTION,
                                              report_items=REPORT_ITEMS,
                                              genome_representation=GENOME_REPRESENTATION,
                                              stats=stats, halloffame=hof, verbose=False)
```
